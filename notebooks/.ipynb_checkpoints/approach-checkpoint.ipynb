```python
data = pd.read_csv('media prediction and its cost.csv')
data.head(2)
```

```python
list(data.columns),len(list(data.columns))
```

```python
data.shape
```

```python
top_view = pd.DataFrame()
for col in list(data.columns):
    
    datum = pd.DataFrame({'col_name':[col],
                          '#uniq_values':[data[col].nunique()],
                          'dtype_':[data[col].dtype],
                          'missing_':[data[col].isna().sum()]})
    top_view = pd.concat([top_view,datum],axis = 0)
    print(f".... analyzing column :{col}")
```

```python
top_view
```

```python
numeric_cols = data.select_dtypes(include = 'number').columns.tolist()
categorical_cols = data.select_dtypes(include = 'object').columns.tolist()
print(f"number of numerical columns :{len(numeric_cols)}")
print(f"number of numerical columns :{len(categorical_cols)}")
```

```python
numeric_cols[0:len(numeric_cols)-1]
```

```python
data[numeric_cols].corr()
```

```python
corr_study = pd.DataFrame()
for col in numeric_cols[0:len(numeric_cols)-1]:
    datum = pd.DataFrame({'col_name':[col],
                          'with_cost':['cost'],
                          'correlation':[data[col].corr(data['cost'])]})
    corr_study = pd.concat([corr_study,datum])
corr_study
```

```python
def corr_between(data,col_name_x,col_name_y):
    corr = data[col_name_x].corr(data[col_name_y])
    return corr

corr_between(data,'store_sales(in millions)','avg_cars_at home(approx)')
```

Food Family Analysis:

```python
categorical_cols
```

```python
numeric_cols
```

```python
data.groupby('food_family').agg(food_depts = pd.NamedAgg('food_department','count'),
                                food_categories = pd.NamedAgg('food_category','count')).reset_index()
```

```python
dat_foodfamily = data.groupby('food_family').agg(food_depts = pd.NamedAgg('food_department','count'),
                                food_categories = pd.NamedAgg('food_category','count'),
                                totalstore_sales = pd.NamedAgg('store_sales(in millions)','sum'),
                                total_store_cost = pd.NamedAgg('store_cost(in millions)','sum'),
                                totalunits_sold = pd.NamedAgg('unit_sales(in millions)','sum')).reset_index()
dat_foodfamily
```

```python
dat_fooddept = data.groupby(['food_family','food_department']).agg(food_categories = pd.NamedAgg('food_category','count'),
                                totalstore_sales = pd.NamedAgg('store_sales(in millions)','sum'),
                                total_store_cost = pd.NamedAgg('store_cost(in millions)','sum'),
                                totalunits_sold = pd.NamedAgg('unit_sales(in millions)','sum')).reset_index()
dat_fooddept
```

```python
dat_foodcat = data.groupby(['food_family','food_department','food_category']).agg(totalstore_sales = pd.NamedAgg('store_sales(in millions)','sum'),
                                total_store_cost = pd.NamedAgg('store_cost(in millions)','sum'),
                                totalunits_sold = pd.NamedAgg('unit_sales(in millions)','sum')).reset_index()
dat_foodcat
```

Media Analysis

```python
media_analysis = data.groupby('media_type').agg(totalsales = pd.NamedAgg('store_sales(in millions)','sum'),
                               total_store_cost = pd.NamedAgg('store_cost(in millions)','sum'),
                               totalunits_sold = pd.NamedAgg('unit_sales(in millions)','sum'),
                               totalcost_aq = pd.NamedAgg('cost','sum')).reset_index()
media_analysis
```
